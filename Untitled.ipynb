```python
import csv
from datetime import datetime
from pprint import pprint
```

```python
from pathlib import Path

files_and_dirs = [p.name for p in Path(".").iterdir() if p.name.endswith('.csv')]
print(files_and_dirs)
```

```python
data_limit = 1000000
data = []
for csvfile in files_and_dirs:
  file = open(csvfile)
  x=1
  for row in file:               #     y     m         d           h       minutes     price
      y, m, d, h, minutes, price = row[:4],row[4:6], row[6:8], row[9:11],row[11:13], row[-14:-3]
      y, m, d, h, minutes = [int(x) for x in [y, m, d, h, minutes]]
      price = float(price)
      date = datetime(y, m, d, h, minutes)
      data.append([date,price])
      #print(row)
      #print(date ,price)
      # ------------------------------- #
      
      if x==data_limit:
          break
      x+=1
```

```python
newdict = {}
for dat in data:
    ymd = dat[0].date()
    time = dat[0].time().hour # select .time() or .time().hour
    price = dat[1]
    if ymd in newdict:
        if newdict[ymd][1] < price:
            newdict[ymd] = [time,price]
    else:
        newdict[ymd] =  [time,price]
    
pprint(newdict)
```

```python
# formatted output
for d in newdict:
  print(f'in {d} the highes price ${newdict[d][1]} was registered at {newdict[d][0]}')
```

```python
hour_points = {}
for d in newdict:
  if newdict[d][0] in hour_points:
    hour_points[newdict[d][0]] += 1
  else:
    hour_points[newdict[d][0]] = 1

pprint(sorted(hour_points.items(), key=lambda x: x[1], reverse=1))
#pprint(newhp)
```
